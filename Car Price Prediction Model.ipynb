```python
df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
df.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python

X = df.drop(['Selling_Price'],axis=1)
Y = df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30,random_state=43)
```

```python
print("Input train shape   :-",x_train.shape)
print("Output train shape :-",y_train.shape)
print("Input test shape     :-",x_test.shape)
print("Output test shape   :-",y_test.shape)
```

```python
## TRAINING MODEL USING LINEAR REGRESSION

from sklearn.linear_model import LinearRegression

MLR_Model = LinearRegression()

MLR_Model.fit(x_train, y_train)
```

```python
print("Coefficent factor :-",MLR_Model.coef_)
print()
print("Intercept :-",MLR_Model.intercept_)
```

```python
Pred_val = MLR_Model.predict(x_test)
```

```python
#    COST

from sklearn.metrics import mean_squared_error
cost = mean_squared_error(y_test,Pred_val)

print("Cost :- ",cost)
```

```python
plt.plot(x_test,y_test, "*", color = "green")
plt.plot(x_test,Pred_val,"*", color = 'orange')
plt.title("Performance Testing")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
